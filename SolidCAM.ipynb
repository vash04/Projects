# Data Acquisition

Doing the data Acquisition.

```python
#importing necessary libraries
import pandas as pd
import numpy as np
```

```python
#loading the dataset and printing first 20 rows
df=pd.read_excel("DATA SAMPLE_Assignment_2.xlsx")
df.head(20)
```

# Data Preprocessing

Cleaning and organizing the dataset extracted from Revenera for effective analysis.

```python
# starting the dataframe from the 10th row
df.columns=df.iloc[10]
df.columns
```

```python
#resetting the df
df=df[11:].reset_index(drop=True)
df
```

De-duplication: 
- Removing duplicate rows to eliminate noise from repeated entries.

```python
df=df.drop_duplicates()
df
```

Focus Columns: 
- Retaining columns like User, Email Domain, Company Domain, Hostname, Egress IP, and Wi-Fi Geolocation for analysis.

```python
focus_columns=['User','Email Domain', 'Company Domain', 'Hostname', 'Egress IP', 'Wifi Location (Latitude)', 'Wifi Location (Longitude)']
```

```python
#data frame focusing specific columns
df = df[focus_columns]
df
```

Handling Missing Values: 
- Replacing Unknown values with NA or defaults and ensuring geolocation fields are numeric.

```python
#replacing unknown with none
df=df.replace({'Unknown': None})
df
```

```python
#ensuring geolocation fields are numeric
df['Wifi Location (Latitude)'] = pd.to_numeric(df['Wifi Location (Latitude)'], errors='coerce')
df['Wifi Location (Longitude)'] = pd.to_numeric(df['Wifi Location (Longitude)'], errors='coerce')
```

```python
#filling null latitude and longitude values as 0 
df=df.fillna({'Wifi Location (Latitude)': 0, 'Wifi Location (Longitude)': 0})
df
```

# Analyzing Commercial Links

Identifying machines linked to commercial entities.

```python
#finding unique values of each column
unique_values = {column: df[column].unique() for column in df.columns}
```

```python
#printing unique value for each column
for column, values in unique_values.items():
    print(f"Column: {column}")
    print(f"Unique Values: {values}")
    print("-" * 50)
```

Filtering Entities: 
- Using keywords like solidcam, caigroup, or others in Email Domain and Company Domain to filter rows.

```python
#setting keywords
keywords = ['solidcam', 'caigroup']
```

```python
#filtering rows
filtered_df = df[
    df['Email Domain'].str.contains('|'.join(keywords), na=False, case=False) |
    df['Company Domain'].str.contains('|'.join(keywords), na=False, case=False)
]
filtered_df
```

Grouping Data: 
- Aggregating data by Hostname and Egress IP to identify frequently occurring machines.

```python
#grouping data by frequency
grouped_df=filtered_df.groupby(['Hostname', 'Egress IP']).size().reset_index(name='Frequency')
grouped_df
```

```python
#grouped data of top 10 Egress IP in order of frequency descending
grouped_df=grouped_df.sort_values(by='Frequency', ascending=False)
grouped_df
```

This is to make a dataframe of WHOIS data

```python
import whois
# Function to perform WHOIS lookup
def get_whois_info(ip):
    try:
        w = whois.whois(ip)
        return w
    except Exception as e:
        return f"Error: {str(e)}"

# List to store WHOIS data
flattened_data = []

# Iterate through unique Egress IPs
unique_ips = grouped_df['Egress IP'].unique()  # Assuming `grouped_df` is already defined
for ip in unique_ips:
    whois_data = get_whois_info(ip)
    if isinstance(whois_data, dict):  # If valid WHOIS data is returned
        # Flatten and store the data
        flattened_data.append({
            "IP": ip,
            "domain_name": whois_data.get("domain_name", "No Data"),
            "creation_date": whois_data.get("creation_date", "No Data"),
            "updated_date": whois_data.get("updated_date", "No Data"),
            "expiration_date": whois_data.get("expiration_date", "No Data"),
            "registrant_organization": whois_data.get("registrant_organization", "No Data"),
            "registrant_address": whois_data.get("registrant_address", "No Data"),
            "country": whois_data.get("country", "No Data"),
        })
    else:
        # Handle errors
        flattened_data.append({
            "IP": ip,
            "domain_name": "Error",
            "creation_date": "Error",
            "updated_date": "Error",
            "expiration_date": "Error",
            "registrant_organization": "Error",
            "registrant_address": "Error",
            "country": "Error",
        })

# Create a DataFrame from the flattened data
df_whois = pd.DataFrame(flattened_data)

# Display the DataFrame
print(df_whois)
```

```python
#whois dataframe
df_whois
```

```python
#merging the dataframe with our filtered dataframe
merged_whois_data=pd.merge(df_whois, filtered_df, left_on="IP", right_on="Egress IP")
merged_whois_data
```

# Wi-Fi Geolocation Analysis

Detecting patterns from geolocation clusters that indicate commercial activity.

Valid Geolocations: 
- Filter out Unknown or invalid coordinates 

```python
#Filtering out invalid coordinates (Latitude and Longitude)
valid_geolocations=filtered_df[
    (filtered_df['Wifi Location (Latitude)'] != 0) &
    (filtered_df['Wifi Location (Longitude)'] != 0) &
    (~filtered_df['Wifi Location (Latitude)'].isnull()) &
    (~filtered_df['Wifi Location (Longitude)'].isnull())
]
valid_geolocations
```

Cluster Analysis: 
- Visualizing clusters of Latitude and Longitude coordinates to find shared locations.

```python
# Extracting Latitude and Longitude
latitude = valid_geolocations['Wifi Location (Latitude)']
longitude = valid_geolocations['Wifi Location (Longitude)']
```

```python
# Scatter plot
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(longitude, latitude, c='blue', alpha=0.5, label='Wi-Fi Location')
plt.title('Wi-Fi Geolocation Clustering')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid()
plt.show()
```

Insights:
- The Majority of our Commercial entities lies in the geolocation cordinate rectangle of the following points : (0,40), (0,46) (20,40), (20,46)

Network Graphs:
- Showing relationships between machines, domains, and IPs using NetworkX.

```python
import networkx as nx
# Create a NetworkX graph
G = nx.Graph()
```

```python
# Adding nodes and edges (example with Egress IPs, Domains, and Machines)
for _, row in filtered_df.iterrows():
    G.add_node(row['Egress IP'], type='IP')
    G.add_node(row['Hostname'], type='Machine')
    G.add_edge(row['Egress IP'], row['Hostname'])  # Creating an edge between IP and Hostname
```

```python
#Visualize the network graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)  # Position the nodes
nx.draw_networkx_nodes(G, pos, node_size=500, node_color="skyblue", alpha=0.8)
nx.draw_networkx_edges(G, pos, edge_color="gray", alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=10, font_color="black")
plt.title("Relationships between Machines, Domains, and IPs")
plt.show()
```

```python
# Filter for top 10 most frequent Egress IPs
top_egress_ips = filtered_df['Egress IP'].value_counts().head(10).index
filtered_data = filtered_df[filtered_df['Egress IP'].isin(top_egress_ips)]
```

```python
# Example: Group Hostnames into categories (e.g., by domain or prefix)
filtered_data['Grouped Hostname'] = filtered_data['Hostname'].apply(lambda x: x.split('.')[0])
```

```python
# Adjusting the layout
pos = nx.spring_layout(G, k=1, seed=42)  # Increased 'k' to spread nodes further apart
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='skyblue', alpha=0.8)
nx.draw_networkx_edges(G, pos, width=1, edge_color='gray', alpha=0.6)
nx.draw_networkx_labels(G, pos, font_size=9, font_color='black')
plt.title("Simplified Network Graph")
plt.show()
```

```python
# Scaled node sizes by frequency
node_sizes = [filtered_df['Egress IP'].value_counts().get(node, 1) * 100 for node in G.nodes]

# Adjusted edge weights (example: uniform or based on frequency)
edge_weights = [2 for edge in G.edges]

# Drawn graph with scaled sizes
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='lightblue', alpha=0.8)
nx.draw_networkx_edges(G, pos, width=edge_weights, edge_color='gray', alpha=0.6)
```

```python
# Create and visualize subgraphs for top Egress IPs
for ip in top_egress_ips:
    subgraph = G.subgraph([ip] + list(filtered_data[filtered_data['Egress IP'] == ip]['Hostname']))
    plt.figure(figsize=(8, 6))
    nx.draw(subgraph, with_labels=True, node_color='lightblue', edge_color='gray')
    plt.title(f"Subgraph for Egress IP: {ip}")
    plt.show()
```

Insights:
- we can see the connection information of Egress IP's to their respective Hostnames by using this graph.
